ndarray型の中身をコピーするには、通常の変数のように`b = a`ではなく、`b = a.copy()`としないといけない。  
単に`b = a`とすると、aの内容の保存先の参照が渡される。`b = a`とした後で`b`の内容を変更すると、その変更が`a`にも反映される。

```python
a = np.array([1, 1])
b = a
print('a = ' + str(a))
print('b = ' + str(b))

b[0] = 100
print('a = ' + str(a))
print('b = ' + str(b))
```

```python
a = np.array([1, 1])
b = a.copy()
print('a = ' + str(a))
print('b = ' + str(b))

b[0] = 100
print('a = ' + str(a))
print('b = ' + str(b))
```

ndarray型だけでなく、list型でも同様の現象が起きる。  
list型の場合には`import copy`としてから、`a = copy.deepcopy()`とすることでコピーできる。

```python
a = [1, 1]
b = a
print('a = ' + str(a))
print('b = ' + str(b))

b[0] = 100
print('a = ' + str(a))
print('b = ' + str(b))
```

```python
import copy

a = [1, 1]
b = copy.deepcopy(a)
print('a = ' + str(a))
print('b = ' + str(b))

b[0] = 100
print('a = ' + str(a))
print('b = ' + str(b))
```

## 2.15 関数

```python
# 関数の定義
def my_func1():
    print('Hi!')

# 関数の実行
my_func1()
```

```python
# 関数の定義
def my_func2(a, b):
    c = a + b
    return c

# 関数の実行
my_func2(1, 2)
```

```python
# 平均値と標準偏差を戻り値とする関数を定義
def my_func3(d):
    m = np.mean(d)
    s = np.std(d)
    return m, s

# 標準正規分布から100個の乱数を生成
data = np.random.randn(100)

# 生成された乱数の平均と標準偏差を取得・出力
data_mean, data_std = my_func3(data)
print(f'mean: {data_mean:.2f}, std: {data_std:.2f}')
```

```python
# 戻り値が複数でも1つの変数で受け取ることができる
output = my_func3(data)
print(output)
print(type(output))
print(f'mean: {output[0]:.2f}, std: {output[1]:.2f}')
```

## 2.16 ファイル保存

1つのndarray型の保存

```python
# データの準備
data = np.array([1, 1, 2, 3, 5, 8, 13])
print(data)
```

```python
# ファイルの保存
np.save('datafile.npy', data)
```

```python
# データの消去
data = []
print(data)
```

```python
# ファイル読み出し
data = np.load('datafile.npy')
print(data)
```

複数のndarray型の保存

```python
# データの準備
data1 = np.array([1, 2, 3])
data2 = np.array([10, 20, 30])
```

```python
# ファイル保存
np.savez('datafile2.npz', data1=data1, data2=data2)
```

```python
# データ消去
data1 = []
data2 = []
```

```python
# ファイル読み出し
outfile = np.load('datafile2.npz')
print(outfile)
```

```python
# 格納している全データの表示
print(outfile.files)
```

```python
# データの取り出し
data1 = outfile['data1']
data2 = outfile['data2']
print(data1)
print(data2)
```
